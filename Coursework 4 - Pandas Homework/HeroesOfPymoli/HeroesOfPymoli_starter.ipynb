```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

current_path = os.getcwd()

# File to Load (Remember to Change These)
file_to_load = current_path + "/data/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load, index_col=0)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
num_unique_items = len(purchase_data['Item ID'].unique())
num_purchases = purchase_data.describe()["Item ID"]["count"]
avg_price = purchase_data.describe()["Price"]["mean"]
total_revenue = purchase_data['Price'].sum()

# Define the data frame
df = pd.DataFrame({
    "Number of Unique Items": [int(num_unique_items)],
    "Average Price": [avg_price],
    "Number of Purchases": [int(num_purchases)],
    "Total Revenue": [total_revenue]
})

df["Average Price"] = df["Average Price"].map('${:,.2f}'.format)
df["Total Revenue"] = df["Total Revenue"].map('${:,.2f}'.format)

df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop the rows with duplicate SN values
df_without_dup = purchase_data.drop_duplicates("SN", "first")

df_new = df_without_dup.loc[:, ["SN", "Gender"]]
df_new_grouped = df_new.groupby("Gender")
df_to_print = df_new_grouped.count()
df_to_print_renamed = df_to_print.rename(columns={"SN": "Total Count"})
df_to_print_renamed["Percentage of Players"] = df_to_print_renamed["Total Count"] / df_to_print_renamed["Total Count"].sum() * 100

df_to_print_renamed["Percentage of Players"] = df_to_print_renamed["Percentage of Players"].map('{:,.2f}'.format)

df_to_print_renamed
```

```python
"""
Another way to solve the above problem, may not be optimal, but worth trying.
"""
# Drop the rows with duplicate SN values
df_without_dup = purchase_data.drop_duplicates("SN", "first")

keys_list = ["female", "male", "other"]
values_list = ["Female", "Male", "Other / Non-Disclosed"]
gender_dict = dict((key, value) for key, value in zip(keys_list, values_list))

count_dict = dict()
percent_dict = dict()

for key, value in gender_dict.items():
    count_dict[key] = df_without_dup["Gender"].value_counts()[value]

count_total = sum(count_dict.values())

for key, value in gender_dict.items():
    percent_dict[key] = count_dict[key] / count_total * 100

df = pd.DataFrame({
    "Total Count": list(count_dict.values()),
    "Percentage of Players": list(percent_dict.values()),
}, index=pd.Index(data=["Female", "Male", "Other / Non-Disclosed"], name='Gender'))

df["Percentage of Players"] = df["Percentage of Players"].map('{:,.2f}'.format)

df  
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [min(df_without_dup.Age) - 1, 9, 14, 19, 24, 29, 34, 39, max(df_without_dup.Age)]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
cats = pd.cut(df_without_dup.Age, bins, labels=labels)
value_counts = cats.value_counts()
percentage_of_players = value_counts / value_counts.sum() * 100

df = pd.DataFrame({
    "Total Count": value_counts,
    "Percentage of Players": percentage_of_players,
}, index=labels)

df["Percentage of Players"] = df["Percentage of Players"].map('{:,.2f}'.format)

df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df_new = purchase_data.loc[:, ["SN", "Price"]]

df_new_grouped = df_new.groupby("SN")
df_to_print = df_new_grouped.count()
df_to_print_renamed = df_to_print.rename(columns={"Price": "Purchase Count"})
df_to_print_renamed["Total Purchase Value"] = df_new_grouped["Price"].sum()
df_to_print_renamed["Average Purchase Price"] = df_to_print_renamed["Total Purchase Value"] / df_to_print_renamed["Purchase Count"]
df_to_print_renamed = df_to_print_renamed[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

df_to_print_renamed.sort_values("Total Purchase Value", ascending=False, inplace=True)

df_to_print_renamed["Average Purchase Price"] = df_to_print_renamed["Average Purchase Price"].map('${:,.2f}'.format)
df_to_print_renamed["Total Purchase Value"] = df_to_print_renamed["Total Purchase Value"].map('${:,.2f}'.format)

df_to_print_renamed.head()
```

```python
"""
Another way to solve the Top Spenders problem, may not be optimal, but worth trying.
"""
SN_list = purchase_data["SN"].unique()

sectors = purchase_data.groupby("SN")

purchase_count_list = []
avg_purchase_price_list = []
total_purchase_value_list = []

for item in SN_list:
    sector_df = sectors.get_group(item)
    purchase_count_list.append(len(sector_df))
    total_purchase_value_list.append(sector_df["Price"].sum())
    avg_purchase_price_list.append(sector_df["Price"].sum() / len(sector_df))

df = pd.DataFrame({
    "Purchase Count": purchase_count_list, 
    "Average Purchase Price": avg_purchase_price_list,
    "Total Purchase Value": total_purchase_value_list,
}, index=pd.Index(data=SN_list, name='SN'))

df.sort_values("Total Purchase Value", ascending=False, inplace=True)

df["Average Purchase Price"] = df["Average Purchase Price"].map('${:,.2f}'.format)
df["Total Purchase Value"] = df["Total Purchase Value"].map('${:,.2f}'.format)

df.head()
```
